## Funcionamiento de los Algoritmos en un Sistema Autónomo de Navegación

### Algoritmo BFS (Búsqueda en Anchura)

En el contexto de un sistema autónomo, como un robot que debe recorrer una red de decisiones o puntos interconectados, el algoritmo BFS permite encontrar el camino más corto desde el punto de partida hasta un destino específico.

Su funcionamiento general es el siguiente:

1. **Inicio**: El sistema selecciona un nodo de origen (punto de partida) y crea una cola vacía. Este nodo inicial es marcado como explorado y se agrega a la cola.

2. **Exploración del entorno**: Mientras existan nodos en la cola, se repiten los siguientes pasos:

   a. Se extrae el siguiente nodo de la cola.  
   
   b. Se procesa este nodo (por ejemplo, registrándolo como parte de la ruta).  
   
   c. Se identifican todos los nodos vecinos no explorados, se los marca como visitados y se los agrega a la cola para su posterior análisis.

3. **Finalización**: Una vez que la cola esté vacía, el sistema habrá explorado todos los caminos posibles accesibles desde el nodo inicial. En este punto, ya se puede reconstruir el camino más corto hasta el objetivo.

Este enfoque es especialmente útil cuando el sistema necesita tomar decisiones rápidas y eficientes para llegar a un destino con el menor número de pasos posibles.

---

### Algoritmo DFS (Búsqueda en Profundidad)

El algoritmo DFS es útil para sistemas autónomos que necesitan explorar profundamente un entorno antes de tomar decisiones, como en casos de inspección completa o búsqueda en entornos desconocidos.

El procedimiento general es:

1. **Inicio**: Se elige un nodo de inicio y se inicializan estructuras para rastrear el recorrido y la profundidad alcanzada.

2. **Exploración recursiva**: El sistema avanza lo más lejos posible desde el nodo actual, moviéndose a los vecinos no visitados. Este proceso se realiza de forma recursiva.

3. **Finalización**: Cuando ya no hay más caminos por explorar, el sistema retrocede y evalúa rutas alternativas. Finalmente, se obtiene el recorrido más profundo posible (el camino más largo hasta el destino, si fue especificado) y la profundidad máxima alcanzada.

Este tipo de exploración es útil para mapear rutas extensas o analizar completamente un entorno antes de actuar.

### 3.2 Comportamiento de los Algoritmos en el Sistema Autónomo

En el diseño de un sistema autónomo capaz de desplazarse a través de una red de puntos de decisión o rutas (como nodos de una red de comunicación o zonas de desplazamiento en un entorno físico), la elección del algoritmo de búsqueda tiene un impacto directo en el comportamiento del agente.

- **BFS (Búsqueda en Anchura)**:  
  Este algoritmo permite al sistema encontrar la **ruta más corta** desde un punto de inicio hasta un destino específico. Es ideal para situaciones en las que se requiere una solución eficiente y rápida, minimizando la cantidad de pasos o decisiones intermedias. BFS evalúa todos los caminos más cercanos antes de expandirse a rutas más largas, por lo que asegura la mejor opción en cuanto a distancia o cantidad de nodos recorridos.

- **DFS (Búsqueda en Profundidad)**:  
  A diferencia del BFS, DFS está diseñado para **explorar tan profundamente como sea posible** antes de retroceder. Esto lo convierte en una herramienta útil en tareas de **inspección exhaustiva**, mapeo completo de entornos o análisis de todas las rutas posibles hacia un objetivo. DFS no garantiza la ruta más corta, pero sí permite obtener una perspectiva completa del entorno antes de tomar una decisión.

Ambos algoritmos fueron implementados sobre un grafo dirigido, lo que significa que las conexiones entre nodos tienen una dirección establecida (de origen a destino). Además, se tomaron precauciones para evitar que el sistema visite múltiples veces los mismos nodos, lo que previene bucles infinitos o decisiones redundantes.

### 3.3 Simulación del Sistema Autónomo

Para validar el comportamiento del sistema, se implementó una simulación en la que un agente autónomo debe desplazarse desde un nodo inicial, como `AS52257`, hasta un nodo de destino, como `AS2914`. Estos nodos pueden representar diferentes entidades dentro de un entorno autónomo, como:

Durante la ejecución, el sistema calcula dos tipos de rutas:

- **Ruta óptima (BFS)**:  
  Se utiliza para determinar el camino más corto desde el origen al destino. Este tipo de ruta es preferido cuando el objetivo del sistema es llegar al punto final de manera directa y eficiente, con el menor número de transiciones posibles.

- **Ruta profunda (DFS)**:  
  Permite al sistema realizar un recorrido más extenso, evaluando posibles rutas largas y descubriendo caminos alternativos o zonas no exploradas. Es especialmente útil en situaciones donde la prioridad no es la velocidad sino la cobertura total del entorno.

Para complementar la simulación, se integró un visualizador desarrollado con las librerías `matplotlib` y `networkx`. Estas herramientas permiten representar gráficamente la estructura del grafo y destacar visualmente las rutas encontradas.

Asuma la aplicación de BFS y DFS en casos de uso reales (por ejemplo: las salidas de evacuación en un campus universitario) y describa el caso de ejemplo o definición del problema y la demo del algoritmo.
```python
from collections import deque
import matplotlib.pyplot as plt
import networkx as nx

# --- Clases ---
class Node:
    """Nodo del grafo identificado por un ID único."""
    def __init__(self, id):
        self.id = id
    def __repr__(self):
        return f"Node({self.id})"
    def __hash__(self):
        return hash(self.id)
    def __eq__(self, other):
        return self.id == other.id

class Arista:
    """Arista dirigida entre dos nodos."""
    def __init__(self, origen, destino):
        self.origen = origen
        self.destino = destino
    def __repr__(self):
        return f"Arista({self.origen.id} -> {self.destino.id})"

class Graph:
    """Grafo dirigido con listas de adyacencia."""
    def __init__(self):
        self.nodos = {}
        self.ady = {}
        self.aristas = []

    def agregar_nodo(self, id_nodo):
        if id_nodo not in self.nodos:
            nodo = Node(id_nodo)
            self.nodos[id_nodo] = nodo
            self.ady[nodo] = []

    def agregar_arista(self, id_origen, id_destino):
        self.agregar_nodo(id_origen)
        self.agregar_nodo(id_destino)
        origen = self.nodos[id_origen]
        destino = self.nodos[id_destino]
        self.ady[origen].append(destino)
        self.aristas.append(Arista(origen, destino))

    def __repr__(self):
        return f"Graph(nodos={len(self.nodos)}, aristas={len(self.aristas)})"

# --- Algoritmos de búsqueda ---
def bfs_shortest_path(graph: Graph, start_id: str, goal_id: str):
    start = graph.nodos.get(start_id)
    goal = graph.nodos.get(goal_id)
    if not start or not goal:
        return None

    queue = deque([start])
    visited = {start}
    predecessor = {start: None}

    while queue:
        current = queue.popleft()
        if current == goal:
            break
        for neighbor in graph.ady[current]:
            if neighbor not in visited:
                visited.add(neighbor)
                predecessor[neighbor] = current
                queue.append(neighbor)
                if neighbor == goal:
                    queue.clear()
                    break

    if goal not in predecessor:
        return None

    path = []
    node = goal
    while node:
        path.append(node.id)
        node = predecessor[node]
    return path[::-1]

def dfs_path(graph: Graph, start_id: str, goal_id: str):
    start = graph.nodos.get(start_id)
    goal = graph.nodos.get(goal_id)
    if not start or not goal:
        return None

    visited = set()
    path = []
    found = False

    def dfs(node):
        nonlocal found
        visited.add(node)
        path.append(node.id)
        if node == goal:
            found = True
            return
        for neighbor in graph.ady[node]:
            if neighbor not in visited and not found:
                dfs(neighbor)
        if not found:
            path.pop()

    dfs(start)
    return path if found else None

# --- Construcción y visualización ---
def construir_grafo():
    g = Graph()
    aristas = [
        ("AS52257", "AS27947"),
        ("AS27947", "AS6762"),
        ("AS6762", "AS3491"), ("AS6762", "AS1299"), ("AS6762", "AS23520"),
        ("AS3491", "AS3356"), ("AS3491", "AS7922"), ("AS3491", "AS6461"),
        ("AS1299", "AS3257"), ("AS1299", "AS6939"), ("AS1299", "AS5511"),
        ("AS1299", "AS174"),  ("AS1299", "AS6453"), ("AS1299", "AS12956"),
        ("AS1299", "AS3320"), ("AS1299", "AS701"),  ("AS1299", "AS7018"),
        ("AS23520", "AS6830"),
        ("AS3356", "AS2914"), ("AS7922", "AS2914"), ("AS6461", "AS2914"),
        ("AS3257", "AS2914"), ("AS6939", "AS2914"), ("AS5511", "AS2914"),
        ("AS174",  "AS2914"), ("AS6453", "AS2914"), ("AS12956","AS2914"),
        ("AS3320", "AS2914"), ("AS701",  "AS2914"), ("AS7018", "AS2914"),
        ("AS6830", "AS2914"),
    ]
    for origen, destino in aristas:
        g.agregar_arista(origen, destino)
    return g

def graficar(g: Graph, ruta, titulo, color):
    G_nx = nx.DiGraph()
    for nodo in g.nodos.values():
        G_nx.add_node(nodo.id)
    for nodo, vecinos in g.ady.items():
        for v in vecinos:
            G_nx.add_edge(nodo.id, v.id)

    pos = nx.spring_layout(G_nx, seed=42)
    plt.figure(figsize=(12, 6))
    nx.draw_networkx_nodes(G_nx, pos, node_size=300, node_color='lightgray')
    nx.draw_networkx_edges(G_nx, pos, edge_color='gray', alpha=0.5)
    nx.draw_networkx_labels(G_nx, pos, font_size=8)

    if ruta:
        aristas_ruta = list(zip(ruta, ruta[1:]))
        nx.draw_networkx_nodes(G_nx, pos, nodelist=ruta, node_color=color, node_size=300)
        nx.draw_networkx_edges(G_nx, pos, edgelist=aristas_ruta, edge_color=color, width=2)

    plt.title(titulo)
    plt.axis('off')

# --- Ejecución Principal ---
def main():
    grafo = construir_grafo()
    print("Nodos disponibles:")
    for i, nodo in enumerate(sorted(grafo.nodos.keys())):
        print(f"{i+1}. {nodo}", end="\t")
        if (i+1) % 5 == 0: print()
    print()

    start = input("Ingresa el nodo de inicio (ejemplo: AS52257): ")
    end = input("Ingresa el nodo de destino (ejemplo: AS2914): ")

    if start not in grafo.nodos or end not in grafo.nodos:
        print("Error: nodo(s) no existente(s) en el grafo.")
        return

    ruta_bfs = bfs_shortest_path(grafo, start, end)
    ruta_dfs = dfs_path(grafo, start, end)

    if ruta_bfs:
        print("Ruta BFS (más corta):", ruta_bfs)
        graficar(grafo, ruta_bfs, f"BFS: Ruta de {start} a {end}", 'green')
    else:
        print(f"No se encontró ruta BFS de {start} a {end}")

    if ruta_dfs:
        print("Ruta DFS (profunda):", ruta_dfs)
        graficar(grafo, ruta_dfs, f"DFS: Ruta de {start} a {end}", 'red')
    else:
        print(f"No se encontró ruta DFS de {start} a {end}")

    if ruta_bfs or ruta_dfs:
        plt.show()

# Ejecutar
main()
```
